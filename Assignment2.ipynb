```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

# Data Cleaning
### Delete Feb 29

```python
df = df[df['Date'].str[-5:]!='02-29']
```

### Transfer temprature to 1 degree

```python
df['Data_Value'] = df['Data_Value']*0.1
```

```python
df
```

### Seperate data into 2 datasets: 2005-2014 and 2015 data

```python
df_14=df[df['Date'].str[:4]!='2015']

print(df_14['Date'].min())
print(df_14['Date'].max())
```

```python
df_15=df[df['Date'].str[:4]=='2015']
print(df_15['Date'].min())
print(df_15['Date'].max())
```

### Delete "Year" from the label

Because we want to draw a line by day of the year

# Get record high/Low temp by day
###### 2005 - 2014

```python
gb_14 = df_14.groupby(['Date','Element'])
df_early = gb_14.agg({'Data_Value' : [min, max]})
df_early.head()
```

```python
df_early.loc[('01-01','TMAX'),('Data_Value','max')]
```

```python
high_14=[]
for x in list(df_early.xs('TMAX',level='Element').xs('max',level=1,axis=1).values):
    high_14.extend(x)
high_14
```

```python
low_14=[]
for x in list(df_early.xs('TMIN',level='Element').xs('min',level=1,axis=1).values):
    low_14.extend(x)
low_14
```

```python
date = list(df_early.index.levels[0])
```

### 2015

```python
gb_15 = df_15.groupby(['Date','Element'])
df_late = gb_15.agg({'Data_Value' : [min, max]})
df_late.head()
```

```python
high_15=[]
for x in list(df_late.xs('TMAX',level='Element').xs('max',level=1,axis=1).values):
    high_15.extend(x)
high_15
```

```python
low_15=[]
for x in list(df_late.xs('TMIN',level='Element').xs('min',level=1,axis=1).values):
    low_15.extend(x)
low_15
```
